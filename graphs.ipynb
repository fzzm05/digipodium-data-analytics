```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainfall  = [17,10,10,10,26,59]


months = ['jan','feb','march','april','may','june']
```

```python

plt.style.use('ggplot')
```

```python
plt.bar(months,rainfall)
```

```python
plt.bar(months,rainfall,color='#ff77b8')

```

```python
plt.bar(months,rainfall)
```

```python
# donut
plt.pie(
    rainfall,
    labels= months,
    autopct ='%1.1f%%',
    
    
    
)
plt.show
```

```python
# donut
plt.pie(
    rainfall,
    labels= months,
    autopct ='%1.1f%%',
    wedgeprops={'width':.4,'edgecolor':'#6666ff','linewidth':2},
    pctdistance=.8
    

    
    
)
plt.show()
```
