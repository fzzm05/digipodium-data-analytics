```python

t =np.arange(-10,10, .2)
y=np.sin(t)
fig,ax =plt.subplots()
ax.plot(t,y,label='sin wave')

ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')
```

```python

t =np.arange(-10,10, .2)
y=np.sin(t)
fig,ax =plt.subplots()
ax.plot(t,y,label='sin wave',color='cyan',linestyle='--',linewidth=3)

ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')
```

```python

t =np.arange(-10,10, .2)
y=np.sin(t)
fig,ax =plt.subplots(figsize=(10,4))
ax.plot(t,y,label='sin wave',color='#4400ff',marker ='h',markersize=5,markerfacecolor='#6600ff',
        alpha=.5)

ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')
```

```python

t =np.arange(-10,10, .2)
y=np.tan(t)
fig,ax =plt.subplots(figsize=(10,4))
ax.plot(t,y,label='sin wave',color='blue',marker ='h',markersize=5,markerfacecolor='red',
        alpha=.5)

ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')
```

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python

data1, data2, data3, data4 = np.random.randn(4, 100) 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```
